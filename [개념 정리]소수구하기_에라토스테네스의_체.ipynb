### **소수 판별 알고리즘 : 어떤 수가 소수인지 판별**

- 1보다 큰 자연수 중에서 1과 자기 자신을 제외한 자연수로는 나누어 떨어지지 않는 자연수



````python
def is_prime_number(x):
  for i in range(2,x): # 2부터 x-1까지의 모든 수를 확인하며
    if x % i == 0:
      return False  # 소수가 아니다
  return True # 소수이다

print(is_prime_number(5))
print(is_prime_number(10))
````

<< 소수 판별 알고리즘의 성능 분석 >>
- 2 ~ X-1까지의 모든 자연수에 대하여 연산을 수행
- 시간 복잡도 : O(x)

> ### **key point! : 약수의 성질**

image.png

````python
# 개선된 코드
# 제곱근을 구하기 위해 math library를 호출
# 2부터 x 제곱근까지의 수로 x를 나눠보고 그 중 나누어 떨어지는 경우가 있다면 소수가 아님
import math

def is_prime_number(x):
  for i in range(2, int(math.sqrt(x))+1):
    if x % i == 0:
      return False
  return True

print(is_prime_number(5))
print(is_prime_number(10))
print(is_prime_number(21))
````

=> 시간 복잡도가 O(x) -> O(x**0.5)로 줄어들었다.

> 파이썬에서 제곱근 구하기

```python

```

2. 연산자 사용

```python

```



### **다수의 소수 판별 : 특정 범위 안에 존재하는 모든 소수 찾기**

- 에라토스테네스의 체 알고리즘을 통해 해결 가능

image.png


- 남아있는 number list의 첫번째 수인 n의 배수를 지우려고 할 때, 입력값으로 들어가는 n은 무조건 소수
- 2 ~ P 사이의 값 중 소수를 찾는 과정에서, 위의 소수 찾기 알고리즘에서도 사용했듯, sqrt(P) 까지의 숫자의 배수만을 전체 number list에서 지우면 그 이후부터는 모두 소수임을 알 수 있다.
image.png

````python
# 에라토스테네스의 체 알고리즘

import math

n = 1000 # 2 ~ 1000까지의 모든 수에 대하여 소수 판별
array = [True for i in range(n+1)]  # 0 ~ n까지 수로 구성된 list 생성
# 처음에는 모든 수가 소수(True)인 것으로 초기화 (0,1 제외)
for i in range(2, int(math.sqrt(n))+1):
  if array[i] == True:
    j = 2
    while i * j <= n:
      array[i*j] = False
      j += 1

#  모든 소수 출력 => 해당 부분이 True인 경우에만 i값이 출력됨
for i in range(2, n+1):
  if array[i]:
    print(i, end = '')
````

<< 에라토스테네스의 체 알고리즘 성능 분석 >>
- 선형 시간에 가까울 정도로 매우 빠르다
- 시간 복잡도 : O(NloglogN)
- **다수의 소수를 찾아야 하는 문제에 효과적으로 사용 가능**
- But, **각 자연수에 대한 소수 여부를 저장해야 하므로 메모리가 많이 필요**
